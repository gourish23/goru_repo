```python
# convert the class column to numeric.
import pandas as pd
df=pd.read_csv("C:\\Users\\gouri\\Desktop\\iris application\\iris.csv",\
              names=['sl','sw','pl','pw','class'])

df['class']=df['class'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
df
```

```python
# generate input and output
x=df.drop(columns=['class'])  # here we have removed class and added in y
y=df['class']
```

```python
x
```

```python
y
```

```python
#split the data set into two parts, case-sensitive so look out for it
1-> one part for training (80%)
2-> other for testing (20%)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train # 80% input , 120 flower are there 
```

```python
x_test # 20% imput, 30 flower
```

```python
 y_train # 80% output, 120 flowers are there
```

```python
y_test # 20% output, 30 flowers are there 
```

```python
# create a machine-learning model KMN

sklearn is package used for machine learning 
KMN -> Knearest Neighbors algorithm


# for a project to be considered good , it should have atleast 90% accuracy 
```

```python
from sklearn.neighbors import KNeighborsClassifier
K=KNeighborsClassifier(n_neighbors=5)

# train the model by 80%  training data
K.fit(x_train,y_train)  # fit is used for giving training to x and y
# test the model by using 30% testing data
y_pred_knn=K.predict(x_test)
y_pred_knn
```

```python
y_test.values # actual ouput
```

```python
# find the accuracy 
from sklearn.metrics import accuracy_score
acc_knn=accuracy_score(y_test,y_pred_knn)
acc_knn=round(acc_knn*100,2)
print("accuracy in KMN model is",acc_knn,'%')
```

```python
# implement logistic regression model-- it imp
# Logistic regression estimates the probability of an event occurring, 
# such as voted or didn't vote, based on a given data set of independent variables.

from sklearn.linear_model import LogisticRegression
L=LogisticRegression()

#train the model
L.fit(x_train,y_train)
# test the model
y_pred_lg=L.predict(x_test)
# find the accuracy
from sklearn.metrics import accuracy_score
acc_lg=accuracy_score(y_test,y_pred_lg)
acc_lg=round(acc_lg*100,2)
print("accuracy in logistic model is",acc_lg,'%')
```

```python
# implement naive bayes model

from sklearn.naive_bayes import GaussianNB
N=GaussianNB()
#train the model
N.fit(x_train,y_train)
# test the model
y_pred_nb=N.predict(x_test)
# find the accuracy
from sklearn.metrics import accuracy_score
acc_nb=accuracy_score(y_test,y_pred_lg)
acc_nb=round(acc_nb*100,2)
print("accuracy in logistic model is",acc_nb,'%')
```

```python
# implement decision tree model

from sklearn.tree import DecisionTreeClassifier
D=DecisionTreeClassifier()
#train the model
D.fit(x_train,y_train)
# test the model
y_pred_dt=D.predict(x_test)
# find the accuracy
from sklearn.metrics import accuracy_score
acc_dt=accuracy_score(y_test,y_pred_dt)
acc_dt=round(acc_dt*100,2)
print("accuracy in tree model is",acc_dt,'%')
```

```python
# compare the accuracy using a bar chart 
import matplotlib.pyplot as plt
models=['KMN','LG','NB','DT']
accuracy=[acc_knn,acc_lg,acc_nb,acc_dt]
plt.bar(models,accuracy,color=['red','green','yellow','orange'])
plt.xlabel("models")
plt.ylabel("accuracy")
plt.show()
```

```python
# convert the class column to numeric.
import pandas as pd
df=pd.read_csv("C:\\Users\\gouri\\Desktop\\iris application\\iris.csv",\
              names=['sl','sw','pl','pw','class'])

df['class']=df['class'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})

# generate input and output
x=df.drop(columns=['class'])  # here we have removed class and added in y
y=df['class']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)

######################################################################################################
# gui for iris project
from tkinter import *
import warnings
warnings.filterwarnings('ignore')
import tkinter.messagebox as m
def kmn():
    global acc_knn
    from sklearn.neighbors import KNeighborsClassifier
    K=KNeighborsClassifier(n_neighbors=5)
    
    # train the model by 80%  training data
    K.fit(x_train,y_train)  # fit is used for giving training to x and y
    # test the model by using 30% testing data
    y_pred_knn=K.predict(x_test)
    # find the accuracy 
    from sklearn.metrics import accuracy_score
    acc_knn=accuracy_score(y_test,y_pred_knn)
    acc_knn=round(acc_knn*100,2)
    m.showinfo(title="KMN",message="accuracy in knn is"+str(acc_knn))
def lg():
    global acc_lg
    global L
    from sklearn.linear_model import LogisticRegression
    L=LogisticRegression()
    
    #train the model
    L.fit(x_train,y_train)
    # test the model
    y_pred_lg=L.predict(x_test)
    # find the accuracy
    from sklearn.metrics import accuracy_score
    acc_lg=accuracy_score(y_test,y_pred_lg)
    acc_lg=round(acc_lg*100,2)
    m.showinfo(title="LG",message="accuracy in lg is"+str(acc_lg))
def nb():
    global acc_nb
    from sklearn.naive_bayes import GaussianNB
    N=GaussianNB()
    #train the model
    N.fit(x_train,y_train)
    # test the model
    y_pred_nb=N.predict(x_test)
    # find the accuracy
    from sklearn.metrics import accuracy_score
    acc_nb=accuracy_score(y_test,y_pred_nb)
    acc_nb=round(acc_nb*100,2)
    m.showinfo(title="NB",message="accuracy in nb is"+str(acc_nb))
def dt():
    global acc_dt
    from sklearn.tree import DecisionTreeClassifier
    D=DecisionTreeClassifier()
    #train the model
    D.fit(x_train,y_train)
    # test the model
    y_pred_dt=D.predict(x_test)
    # find the accuracy
    from sklearn.metrics import accuracy_score
    acc_dt=accuracy_score(y_test,y_pred_dt)
    acc_dt=round(acc_dt*100,2)
    m.showinfo(title="DT",message="accuracy in dt is"+str(acc_dt))
def cmp():
    import matplotlib.pyplot as plt
    models=['KMN','LG','NB','DT']
    accuracy=[acc_knn,acc_lg,acc_nb,acc_dt]
    plt.bar(models,accuracy,color=['red','green','yellow','orange'])
    plt.xlabel("models")
    plt.ylabel("accuracy")
    plt.show()
def submit():
    sl=float(esl.get())
    sw=float(esw.get())
    pl=float(epl.get())
    pw=float(epw.get())
    result=L.predict([[sl,sw,pl,pw]])
    if result[0]==0:
        flower='setosa'
    elif result[0]==1:
        flower='versicolor'
    else: 
        flower='virginica'
    m.showinfo(title="IRIS",message=flower)
def reset():
    esl.delete(0,END)
    esw.delete(0,END)
    epl.delete(0,END)
    epw.delete(0,END)

w=Tk()
w.configure(bg='SlateBlue1')
l1=Label(w,text='IRIS FLOWER PROJECT',bg='azure',font=('arial',20,'bold'),fg='navy',relief="raised")
l1.grid(row=1,column=1,columnspan=4)
###### 2nd row
bkmn=Button(w,bg='azure',text='KMN',font=('arial',20,'bold'),fg='SpringGreen3',command=kmn)
blg=Button(w,bg='azure',text='LG',font=('arial',20,'bold'),fg='SpringGreen3',command=lg)
bnb=Button(w,bg='azure',text='NB',font=('arial',20,'bold'),fg='SpringGreen3',command=nb)
bdt=Button(w,bg='azure',text='DT',font=('arial',20,'bold'),fg='SpringGreen3',command=dt)

bkmn.grid(row=2,column=1)
blg.grid(row=2,column=2)
bnb.grid(row=2,column=3)
bdt.grid(row=2,column=4)
#### 3rd row
bcmp=Button(w,text='COMPARE',bg='azure',font=('arial',20,'bold'),fg='SpringGreen3',command=cmp)
bcmp.grid(row=3,column=2,columnspan=2)
### 4th row
l2=Label(w,text='PREDICT FOR A NEW FLOWER',bg='azure',font=('arial',20,'bold'),fg='navy',relief="raised")
l2.grid(row=4,column=1,columnspan=4)
# 5th row
lsl=Label(w,text='SL',bg='azure',font=('arial',20,'bold'),fg='maroon1')
esl=Entry(w,bg='azure',font=('arial',20,'bold'),fg='PaleVioletRed3')
lsw=Label(w,text='SW',bg='azure',font=('arial',20,'bold'),fg='maroon1')
esw=Entry(w,bg='azure',font=('arial',20,'bold'),fg='PaleVioletRed3')
lsl.grid(row=5,column=1)
esl.grid(row=5,column=2)
lsw.grid(row=5,column=3)
esw.grid(row=5,column=4)

# 6th row
lpl=Label(w,bg='azure',text='PL',font=('arial',20,'bold'),fg='maroon1')
epl=Entry(w,bg='azure',font=('arial',20,'bold'),fg='PaleVioletRed3')
lpw=Label(w,bg='azure',text='PW',font=('arial',20,'bold'),fg='maroon1')
epw=Entry(w,bg='azure',font=('arial',20,'bold'),fg='PaleVioletRed3')

lpl.grid(row=6,column=1)
epl.grid(row=6,column=2)
lpw.grid(row=6,column=3)
epw.grid(row=6,column=4)

# 7th row
bsub=Button(w,bg='azure',text='SUBMIT',font=('arial',20,'bold'),fg='coral1',command=submit)
bres=Button(w,bg='azure',text='RESET',font=('arial',20,'bold'),fg='coral1',command=reset)

bsub.grid(row=7,column=2)
bres.grid(row=7,column=4)

w.mainloop()
```
